# Description of the program
## This program read the vicue data and then seperated it in two groups: G1: 20% data, G2:80% data
## G1 is stored data in a seperad file
## G2 is oversampled by SMOTE and stored data in a seperated file

```python
#Read all data
import csv
import pandas as pd
dataset=pd.read_csv('Dataset/VoiceFeatures.csv')
dataset.head()
```

```python
#estimate the null data in each column
dataset.isnull().sum()
```

```python
#minimum value of each feature
dataset.min()
```

```python
#maimum value of each feature
pd.DataFrame(dataset.max())
```

```python
#normalized the data
df_max_scaled = dataset.copy()
  
# apply normalization techniques on Column 1
for i in range(1,104):
    column = str(i)
    df_max_scaled[column] = df_max_scaled[column] /df_max_scaled[column].abs().max()
    dataset[column]=df_max_scaled[column]
# view normalized data
display(dataset)
```

```python
#replace the user ID by class name and count the number of sample in each class
dataset['Label'] = pd.factorize(dataset['Label'])[0]
dataset.groupby(['Label'])['Label'].count()
```

```python
#seperate the 20.0% data from each profile for test purpose
from sklearn.model_selection import train_test_split
dataset1, test = train_test_split(dataset, test_size=0.2, random_state=42, shuffle=True)
```

```python
#This data will go for oversample
#dataset1['Label'] = pd.factorize(dataset1['Label'])[0]
dataset1.groupby(['Label'])['Label'].count()
```

```python
#store the 20% data in a seperate file
#replace the user ID by class name and count the number of sample in each class
#test['Label'] = pd.factorize(test['Label'])[0] 
test.groupby(['Label'])['Label'].count()
test.to_csv('Dataset\VoiceDatatest.csv')
```

```python
#oversampled the data by using SMOTE
import csv
import numpy as np
import pandas as pd
#import seaborn as sns
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import SMOTE, ADASYN
from collections import Counter

import warnings
warnings.filterwarnings('ignore')

#Special line for normalized and non-normalized data
AllDataSet=dataset
#print(AllDataSet)
#Seperate minority and majority class
length2 = len(AllDataSet)
print(length2)

#indx=0
for indx in range(0,86):
    minClass=AllDataSet[AllDataSet['Label']==indx]
    majClass=AllDataSet[AllDataSet['Label']!=indx]
    minClass["Label"]=0
    majClass["Label"]=1
    majClass=majClass.head(200)
    #print(len(minClass))
    #print(len(majClass))
    frames = [minClass, majClass]
    Class = pd.concat(frames)
    X=Class.drop(['Label'], axis=1)
    Y=Class['Label']
        
    #print(FClass)
    #sns.countplot('ID', data = Class)
    oversample = SMOTE()
    X_res, y_res = oversample.fit_resample(X, Y)

    OversampledData=X_res
    OversampledData['Label']=y_res
    #OversampledData.Label.value_counts()
    dFClass=OversampledData[OversampledData['Label']==0]
    dFClass['Label']=indx
    if indx==0:
        AllClass=dFClass
    else:
        AllClass=pd.concat([AllClass,dFClass])

AllClass.reset_index()    
print(len(AllClass)) 
```
